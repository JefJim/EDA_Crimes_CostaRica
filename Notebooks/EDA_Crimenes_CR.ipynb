```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
sns.set(style='darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Cargar dataset
df = pd.read_csv('../Datasets/crimenes_2020_2021_2022.csv')

# Convertir fechas
df['CRIMEN_FECHA'] = pd.to_datetime(df['CRIMEN_FECHA'], errors='coerce')

# Extraer año, mes, día de la semana
df['AÑO'] = df['CRIMEN_FECHA'].dt.year
df['MES'] = df['CRIMEN_FECHA'].dt.month
df['DIA_SEMANA'] = df['CRIMEN_FECHA'].dt.day_name()
```

```python
df['CRIMEN_TIPO'].value_counts().plot(kind='barh', title="Frecuencia por tipo de crimen")
plt.xlabel("Cantidad de reportes")
plt.ylabel("Tipo de crimen")
plt.tight_layout()
plt.show()
```

```python
crimenes_mensuales = df.groupby(['AÑO', 'MES']).size()
crimenes_mensuales.plot(title= 'Crímenes por mes entre 2020-2022')
plt.ylabel('Cantidad de crímenes')
```

```python
prov =df['LUGAR_PROVINCIA'].value_counts().sort_values()
prov.plot(kind='barh', title = 'Crímenes por provincia')
```

```python
sns.countplot(data=df, x='VICTIMA_GRUPO_ETARIO', order=df['VICTIMA_GRUPO_ETARIO'])
plt.title('Crímenes por grupo etario de la víctima')
```
